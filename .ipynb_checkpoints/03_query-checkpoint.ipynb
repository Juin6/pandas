## loc - 조건필터링

```python
cond = df['who'] == 'man'
cond
```

- case1 : df[condition]

```python
df[cond].head()
```

- case2 : df.loc[condition]

```python
df.loc[cond].head()
```

```python
df[cond]['age'] = 100 # 그냥 바꾸면 바뀌지 않는다.
```

```python
df.loc[cond, 'age'] = 100 # 일괄적으로 어떤 조건에 해당되는 데이터를 바꿀 때에는 location을 이용한다.
```

```python
df.head()
```

```python
cond1 = df['fare'] > 30
cond2 = df['who'] == 'woman'
```

```python
df.loc[cond1 & cond2].head() #cond1 cond2 모두 만족해야 할 경우
```

```python
df.loc[cond1 | cond2].head() #cond1 cond2 중 한개라도 만족할 경우
```

```python
df.iloc[1, 7]
```

```python
df.iloc[[1, 3, 5], [2, 4, 7]]
```

### isin

```python
cond = df['embarked'].isin(['Q', 'S'])
df.loc[cond].head()
```

```python
cond1 = df['embarked'] == 'Q'
cond2 = df['embarked'] == 'S'
df.loc[cond1 | cond2].head()
```

### practice

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
# 남성 흡연자 조건 필터링
cond1 = tips['smoker'] == 'Yes'
cond2 = tips['sex'] == 'Male'
tips.loc[cond1 & cond2].head(10)
```

```python
# tip을 많이 낸 순서로 내림차순 정렬
tips.sort_values('tip', ascending=False)
```

```python
# 상위 10개 출력
tips.sort_values('tip', ascending=False).head(10)
```

```python
#식사인원이 5명 이상인 테이블
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']].head()
```

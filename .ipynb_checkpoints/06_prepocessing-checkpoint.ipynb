```python
df = pd.read_csv('data/seoul_bicycle.csv')
df
```

```python
df.info()
```

```python
df1 = df.copy()
```

```python
df1['대여일자'] = pd.to_datetime(df1['대여일자'], format='%b-%d-%Y')
```

```python
df1['운동량'] = pd.to_numeric(df1['운동량'], errors='coerce')
```

```python
df1.loc[2344]
```

```python
df1.info()
```

```python
df1.loc[2344]
```

```python
df1.tail()
```

```python
bins = [0, 1000, 2000, df1['운동량'].max()]
```

```python
pd.cut(df1['운동량'], bins).value_counts()
```

```python
pd.cut(df1['운동량'], bins, labels=['부족','보통','많음'])
```

```python
df1.describe()
```

```python
df1['운동량_cut'] = pd.cut(df1['운동량'], bins=10) #값을 기준으로 나눠준다.
df1['운동량_cut'].value_counts()
```

```python
df1['운동량_qcut'] = pd.qcut(df1['운동량'], q=10) # 비율을 기준으로 나눠준다.
df1['운동량_qcut'].value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1]
pd.qcut(df1['운동량'], qcut_bins) #출력시 ( 소괄호는 해당 데이터를 포함하지 않는다는뜻, ]대괄호는 해당데이터를 포함한다는 뜻
```

```python
df1['운동량'] = pd.qcut(df1['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
```

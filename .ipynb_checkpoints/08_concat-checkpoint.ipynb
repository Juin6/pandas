### merge

```python
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}
```

```python
money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
info_df = pd.DataFrame(info)
money_df = pd.DataFrame(money)
```

```python
info_df
```

```python
money_df
```

```python
pd.merge(info_df, money_df, left_on='이름', right_on='고객명')
```

```python
pd.merge(info_df, money_df, how='outer', left_on='이름', right_on='고객명')
```

```python
pd.merge(info_df, money_df, left_on='이름', right_on='고객명', how='left')
```

```python
pd.merge(info_df, money_df, how='right', left_on='이름', right_on='고객명')
```

### fillna

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1['age'].fillna(100).tail() #원본 자체 수정이 되지는 않는다.
```

```python
df1.tail()
```

```python
df1['age'] =  df1['age'].fillna(100)
```

```python
df1.tail()
```

```python
df.info()
```

```python
df['deck'].fillna('C')
```

```python
df['deck']
```

```python
# df1['deck'].fillna('No Data')
# => 카테고리타입은 카테고리에 없는 정보는 넣을 수 없음
```

```python
df1['deck'] = df1['deck'].cat.add_categories('No Data')
df1['deck'].fillna('No Data')
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
```

```python
df1['age'] = df1['age'].fillna(mean_age)
df1['age'].tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

- 남자 승객의 결측치 => 남자 승객의 평균
- 여자 승객의 결측치 => 여자 승객의 평균

```python
cond1 = df1['sex'] == 'male'
mean1 = df1.loc[cond1, 'age'].mean()

cond2 = df1['sex'] == 'female'
mean2 = df1.loc[cond2, 'age'].mean()
```

```python
df1
```

```python
mean1
```

```python
mean2
```

```python
df1.loc[cond1, 'age'] = df1.loc[cond1, 'age'].fillna(mean1)
```

```python
df1.loc[cond2, 'age'] = df1.loc[cond2, 'age'].fillna(mean2)
```

```python
df1['age'].mean()
```

```python
c1 = df1['sex'] == 'male'
df.loc[c1, 'age']
```
